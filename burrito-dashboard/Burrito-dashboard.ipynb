```python
import pandas as pd
import numpy as np
import atoti as tt
import re
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('burritos_01022018.csv')
```

```python
df.head
```

```python
rm_par = df.columns = [re.sub("([\(\[]).*?([\)\]])", "", x).strip() for x in df.columns]
rm_col = df.columns = [x.replace(':','_').strip() for x in df.columns]
```

```python
df.isnull().sum()/df.shape[0]
```

```python
session = tt.Session()
```

```python
scaler = MinMaxScaler()
```

```python
burrito_vars_norm = df.loc[:, ['Circum', 'Volume', 'Length', 'Mass', 'Cost']]
```

```python
burrito_vars_norm.isnull().sum()
```

```python
bnorms = scaler.fit_transform(burrito_vars_norm)*10
```

```python
df[['Circum_norm', 'Volume_norm', 'Length_norm', 'Mass_norm', 'Cost_norm']] = bnorms
```

```python
df.columns
```

```python
burrito_variables = pd.melt(df.reset_index(), id_vars=['Location', 'Burrito'],
                            value_vars=['Circum_norm', 'Volume_norm', 'Length_norm', 'Mass_norm', 'Cost_norm'])
```

```python
burrito_variables
```

```python
burrito_table = session.read_pandas(df, table_name='burritos')
```

```python
burrito_table.head()
```

```python
cube = session.create_cube(burrito_table)
```

```python
h = cube.hierarchies
l = cube.levels
m = cube.measures
```

```python
m['five'] = 5
m['lenXwrap'] = m['Length.MEAN']*m['Wrap.MEAN']
```

```python
m['five']
```

```python
session.visualize('exploration 1')
```

```python
session.visualize('scatter plot nb-yelp-google')
```

```python
burrito_var_table = session.read_pandas(burrito_variables, table_name = 'burrito_variables', keys = ['Location', 'Burrito', 'variable'])
```

```python
burrito_table.join(burrito_var_table)
```

```python
m['value']=tt.agg.mean(burrito_var_table['value'])
m['aggvalue'] = tt.agg.mean(m['value'], scope = tt.scope.origin(l['Location'], l['Burrito'], l['variable'])) 
```

```python
session.visualize('radar chart final')
```
